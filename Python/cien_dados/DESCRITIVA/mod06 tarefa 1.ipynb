####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python


# Criar um gráfico de barras para a variável APGAR1
apgar1_counts = sinasc['APGAR1'].value_counts()
plt.bar(apgar1_counts.index, apgar1_counts.values)
plt.title('Distribuição de APGAR1')
plt.xlabel('APGAR1')
plt.ylabel('Número de Nascimentos')
plt.show()

# Criar um gráfico de barras para a variável APGAR5
apgar5_counts = sinasc['APGAR5'].value_counts()
plt.bar(apgar5_counts.index, apgar5_counts.values)
plt.title('Distribuição de APGAR5')
plt.xlabel('APGAR5')
plt.ylabel('Número de Nascimentos')
plt.show()

# Criar um gráfico de barras para a variável SEXO
sexo_counts = sinasc['SEXO'].value_counts()
plt.bar(sexo_counts.index, sexo_counts.values)
plt.title('Distribuição de Sexo')
plt.xlabel('Sexo')
plt.ylabel('Número de Nascimentos')
plt.show()

# Criar um gráfico de barras para a variável GESTACAO
gestacao_counts = sinasc['GESTACAO'].value_counts()
plt.bar(gestacao_counts.index, gestacao_counts.values)
plt.title('Distribuição de Gestação')
plt.xlabel('Gestação')
plt.ylabel('Número de Nascimentos')
plt.tick_params(axis="x", rotation = 90)
plt.show()

# Criar um gráfico de barras para a variável ESTCIVMAE
estcivmae_counts = sinasc['ESTCIVMAE'].value_counts()
plt.bar(estcivmae_counts.index, estcivmae_counts.values)
plt.title('Distribuição de Estado Civil da Mãe')
plt.xlabel('Estado Civil da Mãe')
plt.ylabel('Número de Nascimentos')
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
apgar5_por_consultas = sinasc.groupby('CONSULTAS')['APGAR5'].mean()
apgar5_por_consultas
```

```python
plt.bar(apgar5_por_consultas.index, apgar5_por_consultas.values)
plt.title('Distribuição de APGAR1')
plt.xlabel('CONSULTAS')
plt.ylabel('Número de Nascimentos')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Configurar o layout dos subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))


# Criar um gráfico de barras para a variável SEXO
sexo_counts = sinasc['SEXO'].value_counts()
axs[0, 0].bar(sexo_counts.index, sexo_counts.values)
axs[0, 0].set_title('Distribuição de Sexo')

# Criar um gráfico de barras para a variável GESTACAO
gestacao_counts = sinasc['GESTACAO'].value_counts()
axs[0, 1].bar(gestacao_counts.index, gestacao_counts.values)
axs[0, 1].set_title('Distribuição de Gestação')
axs[0, 1].tick_params(axis="x", rotation = 90)


# Criar um gráfico de barras para a variável ESTCIVMAE
estcivmae_counts = sinasc['ESTCIVMAE'].value_counts()
axs[1, 0].bar(estcivmae_counts.index, estcivmae_counts.values)
axs[1, 0].set_title('Distribuição de Estado Civil da Mãe')

consultas_counts = sinasc['CONSULTAS'].value_counts()
axs[1, 1].bar(consultas_counts.index, consultas_counts.values)
axs[1, 1].set_title('Distribuição de CONSULTAS')

# Configurar a aparência geral do gráfico
plt.subplots_adjust(wspace=0.3, hspace=0.5)
plt.suptitle('Distribuição de Variáveis no SINASC de RO em 2019', fontsize=16)
plt.show()
```

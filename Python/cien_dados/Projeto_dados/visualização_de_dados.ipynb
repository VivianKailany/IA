```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid")
```

```python
df=pd.read_csv('2. Python_M10_support material.csv', na_values='na')
df.head()
```

```python
fn = lambda valor: float(valor.replace(".","").replace(",","."))
```

```python
df['valor_transacoes_12m'] = df['valor_transacoes_12m'].apply(fn)
```

```python
df['limite_credito'] = df['limite_credito'].apply(fn)
```

```python
df.dropna(inplace=True) #remove lnhas problematicas
```

```python
f = sns.relplot(data=df, x='valor_transacoes_12m', y='qtd_transacoes_12m', hue='default')
f=f.set(title='Relao entre valor e quantidade de transacoes no ultimo',
      xlabel = 'valor das transaes no ltimo ano',
      ylabel = 'Quantidade das Transaes noltimo ano'
     )
```
